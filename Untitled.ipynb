```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:\\Users\\Ramzan\\Desktop\\Sparks Foundation\\TASK2\\SampleSuperstore.csv")
```

```python
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
data.hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms of Numerical Variables", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
```

```python
plt.subplot(2, 2, 1)
sns.countplot(x='Category', data=data)
```

```python
plt.subplot(2, 2, 2)
sns.countplot(x='Region', data=data)
```

```python
plt.subplot(2, 2, 3)
sns.countplot(x='Sub-Category', data=data, hue='Category')
plt.xticks(rotation=90)
```

```python
plt.suptitle("Categorical Variables Exploration", y=1.02)
plt.show()
```

```python
profit_by_region = data.groupby('Region')['Profit'].sum().sort_values()
plt.figure(figsize=(10, 6))
profit_by_region.plot(kind='bar')
plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```
